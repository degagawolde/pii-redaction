```python
from google import genai
from google.genai import types

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
client = genai.Client()
contents = """
 personally identifiable information(PII) includes:
- Human names (including partial names, nicknames)
- Company names
- Dates (various formats: MM/DD/YYYY, DD-MM-YYYY, written dates)
- Addresses (street addresses, cities, postal codes)
- Email addresses
- Phone numbers

 Text: 
 My name is Degaga wolde. You can call me on 0947381036 or email me on degagawolde@gmail.com."""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=contents,
    config=types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(thinking_budget=0),  # Disables thinking
        system_instruction="Identify and list the PII, in the given text",
    ),
)
print(response.text)
```
